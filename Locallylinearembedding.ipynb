```python
# Step 2: Combine both datasets
combined = pd.concat([train, test], ignore_index=True)
```

```python
# Step 3: Split features and labels
X = combined.drop(columns=['subject', 'Activity'])
y = combined['Activity']
```

```python
import pandas as pd
from sklearn.manifold import LocallyLinearEmbedding
import matplotlib.pyplot as plt
import seaborn as sns

# Load combined dataset (assuming already uploaded as train.csv + test.csv)
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
combined_data = pd.concat([train, test], ignore_index=True)

# Focus on WALKING vs WALKING_UPSTAIRS
subset_walk = combined_data[combined_data['Activity'].isin(['WALKING', 'WALKING_UPSTAIRS'])]
X_walk = subset_walk.drop(columns=['subject', 'Activity'])
y_walk = subset_walk['Activity']

# Apply LLE
lle_walk = LocallyLinearEmbedding(n_components=2, n_neighbors=15)
X_lle_walk = lle_walk.fit_transform(X_walk)

# Plot
lle_df_walk = pd.DataFrame(X_lle_walk, columns=["LLE1", "LLE2"])
lle_df_walk["Activity"] = y_walk.values

plt.figure(figsize=(10, 6))
sns.scatterplot(data=lle_df_walk, x="LLE1", y="LLE2", hue="Activity", palette="Set1")
plt.title("LLE: WALKING vs WALKING_UPSTAIRS (k=15)")
plt.xlabel("LLE Component 1")
plt.ylabel("LLE Component 2")
plt.legend(title="Activity")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE, LocallyLinearEmbedding
import matplotlib.pyplot as plt
import seaborn as sns


# Use only a subset of features (e.g., tBodyAcc-mean()-X/Y/Z)
selected_features = ['tBodyAcc-mean()-X', 'tBodyAcc-mean()-Y', 'tBodyAcc-mean()-Z']
X = combined[selected_features]
y = combined['Activity']

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
df_pca = pd.DataFrame(X_pca, columns=["Dim1", "Dim2"])
df_pca["Method"] = "PCA"
df_pca["Activity"] = y

# t-SNE
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
X_tsne = tsne.fit_transform(X_scaled)
df_tsne = pd.DataFrame(X_tsne, columns=["Dim1", "Dim2"])
df_tsne["Method"] = "t-SNE"
df_tsne["Activity"] = y

# LLE
lle = LocallyLinearEmbedding(n_neighbors=30, n_components=2)
X_lle = lle.fit_transform(X_scaled)
df_lle = pd.DataFrame(X_lle, columns=["Dim1", "Dim2"])
df_lle["Method"] = "LLE"
df_lle["Activity"] = y

# Combine all
df_all = pd.concat([df_pca, df_tsne, df_lle], ignore_index=True)

# Plot
plt.figure(figsize=(18, 5))
sns.set(style="whitegrid")
plot = sns.scatterplot(data=df_all, x="Dim1", y="Dim2", hue="Activity", style="Method", palette="Set2", alpha=0.7)
plt.title("Comparison of PCA, t-SNE, and LLE using tBodyAcc-mean()-X/Y/Z")
plt.xlabel("Component 1")
plt.ylabel("Component 2")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
